```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.style.use('ggplot') 
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler,StandardScaler,OneHotEncoder
from sklearn.model_selection import cross_val_score,train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.metrics import confusion_matrix
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
X  = pd.read_csv(r'..\data\Data Science London\train.csv',header=None)
y = pd.read_csv(r'..\data\Data Science London\trainLabels.csv',header=None)
df_test = pd.read_csv(r'..\data\Data Science London\test.csv',header=None)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#特徵縮放
scaler = StandardScaler() 
X_train= scaler.fit_transform(X_train)

#baseline  


LR_model = LogisticRegression(random_state = 0)
print(cross_val_score(LR_model,X_train,y_train).mean())

GB = GradientBoostingClassifier(random_state=0)
print(cross_val_score(GB,X_train,y_train).mean())

RF_model = RandomForestClassifier(random_state = 0)
print(cross_val_score(RF_model,X_train,y_train).mean())

# 進行訓練
XGB_model = XGBClassifier(random_state = 0)
print(cross_val_score(XGB_model,X_train,y_train).mean())
```

```python
X_tmp = scaler.fit_transform(X)

LR = LogisticRegression(random_state = 0)
print(cross_val_score(LR,X_tmp,y).mean())

GB = GradientBoostingClassifier(random_state=0)
print(cross_val_score(GB,X_tmp,y).mean())

RF = RandomForestClassifier(random_state = 0)
print(cross_val_score(RF,X_tmp,y).mean())

# 進行訓練
XGB = XGBClassifier(random_state = 0)
print(cross_val_score(XGB,X_tmp,y).mean())
```

## 調整最佳超參數

```python
# 找尋最佳參數的函示
def find_bestparam(alg,X,y):
    n_estimators = [50, 100, 150,200]
    max_depth = [1, 3, 5,7,9]
    param_grid = dict(n_estimators=n_estimators, max_depth=max_depth)

    ## 建立搜尋物件，放入模型及參數組合字典 (n_jobs=-1 會使用全部 cpu 平行運算)
    grid_search = GridSearchCV(GB, param_grid, scoring="neg_mean_squared_error", n_jobs=-1, verbose=1)

    # 開始搜尋最佳參數
    grid_result = grid_search.fit(X, y)

    # 使用最佳參數重新建立模型
    alg_bestparam =  alg(max_depth=grid_result.best_params_['max_depth'],
                                               n_estimators=grid_result.best_params_['n_estimators'])
    
    score = cross_val_score(alg_bestparam,X,y).mean()
    # 印出最佳結果與最佳參數
    print ("Best cross_val_score: %f using %s" % (score, grid_result.best_params_))
```

```python
alg_list = [GradientBoostingClassifier, RandomForestClassifier, XGBClassifier]
for alg in alg_list:
    find_bestparam(alg,X,y)


```
